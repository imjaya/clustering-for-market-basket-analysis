## Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn import mixture
```

## Importing Dataset

```python
# Importing the dataset
dataset = pd.read_csv('Mall_Customers.csv')
df = dataset.copy()

# Visualizing data
df.head()



```

## Encoding 

```python


# One Hot Encoding the categorical data - Gender
df = pd.get_dummies(df, columns = ['Gender'], prefix = ['Gender'])


df.head()
# Please run the previous cell again if an error shows up
```

## Picking Features 0 --> Male and 1--> Female 

```python

X = df.iloc[:, [1,2,3,4]].values
print(X)
```

# K-Means Implementation

## PCA Of The Data For Plotting

```python


ndimensions = 2
X1=X
pca = PCA(n_components=ndimensions)
pca.fit(X1)
X_pca_array = pca.transform(X1)
```

## K-means and Plotting For K=2

```python

kmeans2 = KMeans(n_clusters = 2)
kmeans2.fit(X)

labels_k2=kmeans2.labels_
centroid2=kmeans2.cluster_centers_
```

```python
labels_k2=np.array(labels_k2)
labels_k2=labels_k2+1

fig, ax = plt.subplots(figsize = (7,7))
scatter=ax.scatter(X_pca_array[:, 0], X_pca_array[:, 1],c=labels_k2,cmap='rainbow')

pca.fit(centroid2)
pca_mean = pca.transform(centroid2)
plt.scatter(pca_mean[:, 0], pca_mean[:, 1], c='black', s=200, alpha=0.8);

legend = ax.legend(*scatter.legend_elements(), title="Clusters")
ax.add_artist(legend)
                            
ax.set_xlabel('PC-1')
ax.set_ylabel('PC-2')
ax.grid()
ax.axhline(0, color='black')
ax.axvline(0, color='black')
ax.set_title("Principal Components Analysis of Dataset Under Kmeans For K=2");
```

## Elbow Plot Of Clusters To Decide The Best Fit

```python

Cost_Function = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(X)
    Cost_Function .append(kmeans.inertia_)


#Plotting Number of Clusters Vs Cost Function - The Elbow Method

plt.plot(range(1, 11), Cost_Function )
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Cost Function')
plt.show()
```

## BIC Penalty Plot For GMM

```python
from sklearn import mixture
n_components_range = range(1, 11)
lowest_bic = np.infty
bic = []
for n_comp in n_components_range:
    # Fit a Gaussian mixture with EM
    gmm = mixture.GaussianMixture(n_components=n_comp,
                                  covariance_type='full')
    gmm.fit(X)
    bic.append(gmm.bic(X))
    if bic[-1] < lowest_bic:
        lowest_bic = bic[-1]
        best_gmm = gmm

bic = np.array(bic)
```

```python
plt.plot(range(1, 11), bic)
plt.title('The BIC Penalty Argument')
plt.xlabel('Number of clusters')
plt.ylabel('BIC')
plt.show()
```
